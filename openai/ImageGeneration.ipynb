# 图像生成（DALL·E）

```python
import openai
from IPython.display import Image
```

## 创建

```python
response = openai.Image.create(
  prompt="画一只军舰鸟",
  n=1,
  size="1024x1024"
)
image_url = response['data'][0]['url']

from IPython.display import Image
Image(url=image_url)
```

```python
def create_image(prompt, size="1024x1024"):
  response = openai.Image.create(
    prompt=prompt,
    n=1,
    size=size
  )
  image_url = response['data'][0]['url']

  return Image(url=image_url, width=512, height=512)
```

```python
create_image("画一只猫和狗")
```

```python
response = openai.Image.create(
  prompt="桌子上是儿子14岁的生日蛋糕，上面点亮了蜡烛，桌子上还摆满了丰盛的菜。",
  n=1,
  size="1024x1024"
)
image_url = response['data'][0]['url']
print(image_url)

from IPython.display import Image
Image(url=image_url)
```

```python
response = openai.Image.create(
  prompt="On the table is the son's 14th birthday cake, lit with candles, and the table is full of sumptuous dishes.",
  n=1,
  size="1024x1024"
)
image_url = response['data'][0]['url']
print(image_url)

from IPython.display import Image
Image(url=image_url)
```

```python
response = openai.Image.create(
  prompt="一只红色的卡通狗头的LOGO",
  n=4,
  size="512x512"
)

for i in range(0, len(response['data'])):
  image_url = response['data'][i]['url']
  display(Image(url=image_url))
```
